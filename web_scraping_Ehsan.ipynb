```python
site = 'https://www.ehsan.sa/tyassarat'
# Launch the browser
driver = webdriver.Chrome(chromedriver)
#driver = webdriver.Chrome(ChromeDriverManager().install())

# open the booking home page 
driver.get(site)
```

```python
soup= BeautifulSoup(driver.page_source,'html.parser')
soup
```

```python
num_page=60

ID = []
case = []
debt = []
link = []
percentage = []
```

```python
for i in range(num_page):
    
    soup= BeautifulSoup(driver.page_source,'html.parser')
    for index, t in enumerate(soup.find_all("div",class_="col-md-6 col-lg-4")):
        ua = UserAgent()
        user_agent = {'User-agent': ua.random}
        print(user_agent)

        response  = requests.get(site, headers = user_agent,  verify = False)
        print("status:", response.status_code)
        if response.status_code != 200:
            continue

        ID.append(t.find("small", class_="tanfeeth-bill bg-white text-black-555 border-rounded-8 p-2").span.text)
        case.append(t.find("div",class_="text-white").text)
        debt.append(t.find("div", class_="text-black-555 font-semibold").text)
        link.append(t.find("button", class_="btn btn-primary-blue rounded-pill px-3 w-100 py-1 prevent-click-onscroll")["onclick"].replace("location.href=", ""))   
        percentage.append(t.find("div", class_="donut-center").text) 
    time.sleep(5)
    Next_button=driver.find_element_by_xpath('//li[contains(@class,"page-item  active")]/following-sibling::li') 
    Next_button.click()
    time.sleep(7)
    if(i+1%20==0):
        time.sleep(90)# after 20 page waiting 2 minutes

    
```

```python
df_main = pd.DataFrame(
    {'ID': ID,
     'Case': case,
     'Debt': debt, 
     'Link': link,
     'Percentage':percentage
    }) 


df_main
```

```python
l = df_main.Link[0]
```

```python
#from fake_useragent import UserAgent
soup_list =[]

for l in df_main.Link:
    base_url = 'https://www.ehsan.sa'
    url = base_url + l
    site = url.replace("'", "")


    ua = UserAgent()
    user_agent = {'User-agent': ua.random}
    print(user_agent)

    response  = requests.get(site, headers = user_agent,  verify = False)
    print("status:", response.status_code)
    if response.status_code != 200:
        continue

    soup_list.append(BeautifulSoup(response.text))
    time.sleep(3)

    if (i+1 % 20 == 0):
        time.sleep(90)
```

```python
# here 
subsite_dict = {
    "vistors" : [],
    "last_donation":[],
    "num_donation":[]
}
for soup_index, local_soup in enumerate(soup_list):
    elements = local_soup.find_all("span", class_ = "text-primary-green h5 m-0 text-nowrap")
    if len(elements) == 3:
        for keys, local_block in zip(["vistors","last_donation","num_donation"], elements) :
            subsite_dict[keys].append(local_block.text)

df_new = pd.DataFrame(subsite_dict)
df_main = df_main.merge(df_new, left_index=True, right_index=True)
```
